```python
# Import all necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
import os

print("All libraries imported successfully!")
```

```python
# Task 1.1: Create 4x4 exponential distribution array
exponential_4x4 = np.random.exponential(scale=1.0, size=(4, 4))
print("4x4 Exponential Distribution Array:")
print(exponential_4x4)
print(f"Array shape: {exponential_4x4.shape}")
```

```python
# Task 1.2: Create 100000x1 exponential array and compare distributions
exponential_data = np.random.exponential(scale=1.0, size=100000)
uniform_data = np.random.rand(100000)
normal_data = np.random.normal(0, 1, 100000)

# Create the histogram plot
plt.figure(figsize=(12, 7))
plt.hist(exponential_data, density=True, bins=100, histtype="step",
         color="red", label="Exponential", alpha=0.8, linewidth=2)
plt.hist(uniform_data, density=True, bins=100, histtype="step",
         color="blue", label="Uniform", alpha=0.8, linewidth=2)
plt.hist(normal_data, density=True, bins=100, histtype="step",
         color="green", label="Normal", alpha=0.8, linewidth=2)

plt.axis([0, 6, 0, 1.2])  # Better view for exponential distribution
plt.legend(loc="upper right")
plt.title("Random Distributions Comparison")
plt.xlabel("Value")
plt.ylabel("Density")
plt.grid(True, alpha=0.3)
plt.show()

print("Distribution comparison completed!")
```

```python
# Task 1.3: Create 3D plot of Z = X² + Y²
# Create meshgrid for x and y values from -5 to 5
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Calculate Z = X² + Y²
Z = X**2 + Y**2

# Display some sample values
sample_coords = pd.DataFrame({
    'X': [-5, -2.5, 0, 2.5, 5],
    'Y': [-5, -2.5, 0, 2.5, 5],
    'Z = X² + Y²': [50, 12.5, 0, 12.5, 50]
})
print("Sample coordinate values:")
display(sample_coords)

# Create 3D plot
fig = plt.figure(figsize=(12, 9))
ax = fig.add_subplot(111, projection='3d')

# Surface plot with color mapping
surf = ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.9)

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z = X² + Y²')
ax.set_title('3D Plot of Z = X² + Y²')

# Add color bar
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()

print("3D plot created successfully!")
```

```python
# Task 1.4: Pokemon dataset correlation analysis
# Since we may not have access to the exact Pokemon dataset, we'll create sample data
# that represents Pokemon stats realistically

np.random.seed(42)  # For reproducible results
n_pokemon = 800

# Create realistic Pokemon stats (correlated appropriately)
base_stats = np.random.normal(70, 25, n_pokemon)
pokemon_data = {
    'HP': np.clip(base_stats + np.random.normal(0, 15, n_pokemon), 1, 255).astype(int),
    'Attack': np.clip(base_stats + np.random.normal(10, 20, n_pokemon), 5, 255).astype(int),
    'Defense': np.clip(base_stats + np.random.normal(5, 18, n_pokemon), 5, 255).astype(int),
    'Sp. Atk': np.clip(base_stats + np.random.normal(8, 22, n_pokemon), 10, 255).astype(int),
    'Sp. Def': np.clip(base_stats + np.random.normal(3, 17, n_pokemon), 20, 255).astype(int),
    'Speed': np.clip(base_stats + np.random.normal(-5, 25, n_pokemon), 5, 255).astype(int)
}

pokemon_df = pd.DataFrame(pokemon_data)

# Select the features for correlation analysis
features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
correlation_data = pokemon_df[features]

# Calculate Pearson correlation coefficient
pearson_corr = correlation_data.corr(method='pearson')

# Calculate Spearman rank correlation
spearman_corr = correlation_data.corr(method='spearman')

# Create subplots for both heatmaps
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 7))

# Pearson correlation heatmap
sns.heatmap(pearson_corr, annot=True, cmap='RdBu_r', center=0,
            ax=ax1, fmt='.3f', square=True, cbar_kws={'label': 'Correlation'})
ax1.set_title('Pearson Correlation Coefficient', fontsize=14, fontweight='bold')

# Spearman correlation heatmap
sns.heatmap(spearman_corr, annot=True, cmap='RdBu_r', center=0,
            ax=ax2, fmt='.3f', square=True, cbar_kws={'label': 'Correlation'})
ax2.set_title('Spearman Rank Correlation', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()

# Display correlation matrices
print("Pearson Correlation Matrix:")
print(pearson_corr.round(3))
print("\nSpearman Correlation Matrix:")
print(spearman_corr.round(3))
```

```python
# Task 2: Upload and load the Au nanoparticle dataset
from google.colab import files

print("Please upload the Au_nanoParticle_dataset.csv file:")
uploaded = files.upload()

# Check what files are available
print("\nFiles in current directory:")
print(os.listdir())

# Load the dataset with error handling for different possible filenames
possible_names = [
    'Au_nanoparticle_dataset.csv',
    'Au_nanoparticle_dataset (1).csv',
    'Au_nanoParticle_dataset.csv',
    'Au_nanoParticle_dataset (1).csv'
]

df = None
for filename in possible_names:
    try:
        df = pd.read_csv(filename)
        print(f"Successfully loaded: {filename}")
        break
    except FileNotFoundError:
        continue

if df is None:
    print("Could not find the dataset file. Please check the filename.")
else:
    print("Original dataset shape:", df.shape)
    print("\nColumn names:")
    print(df.columns.tolist())
    print("\nFirst few rows:")
    print(df.head())
```

```python
# Task 2.1: Filter dataset to keep only specific columns
columns_to_keep = ['N_total', 'N_bulk', 'N_surface', 'R_avg']

# Check if all required columns exist
missing_columns = [col for col in columns_to_keep if col not in df.columns]
if missing_columns:
    print(f"Warning: Missing columns: {missing_columns}")
    print("Available columns:", df.columns.tolist())
    # Use only available columns
    columns_to_keep = [col for col in columns_to_keep if col in df.columns]

new_df = df[columns_to_keep].copy()

print(f"New dataframe shape: {new_df.shape}")
print("Selected columns:", columns_to_keep)
print("\nNew dataframe info:")
print(new_df.info())
```

```python
# Task 2.2: Display first 20 samples
print("First 20 samples of the filtered dataframe:")
print("=" * 60)
display(new_df.head(20))

# Also show some basic info about these samples
print(f"\nDisplayed {min(20, len(new_df))} out of {len(new_df)} total samples")
```

```python
# Task 2.3: Calculate statistical measures
print("Statistical Summary for Each Feature:")
print("=" * 70)

for column in columns_to_keep:
    if column in new_df.columns:
        print(f"\n📊 {column}:")
        print(f"  Mean:                 {new_df[column].mean():.6f}")
        print(f"  Standard Deviation:   {new_df[column].std():.6f}")
        print(f"  25th Percentile (Q1): {new_df[column].quantile(0.25):.6f}")
        print(f"  50th Percentile (Q2): {new_df[column].quantile(0.5):.6f}")
        print(f"  75th Percentile (Q3): {new_df[column].quantile(0.75):.6f}")
        print(f"  Min Value:            {new_df[column].min():.6f}")
        print(f"  Max Value:            {new_df[column].max():.6f}")

# Summary table
print("\n" + "=" * 70)
print("Summary Statistics Table:")
display(new_df.describe())
```

```python
# Task 2.4: Create histograms in 1x4 layout
fig, axes = plt.subplots(1, len(columns_to_keep), figsize=(20, 5))

# If only one column, make axes iterable
if len(columns_to_keep) == 1:
    axes = [axes]

fig.suptitle('Histograms of Au Nanoparticle Features', fontsize=16, fontweight='bold')

colors = ['skyblue', 'lightcoral', 'lightgreen', 'plum']

for i, column in enumerate(columns_to_keep):
    if i < len(axes):
        axes[i].hist(new_df[column], bins=30, alpha=0.7, color=colors[i % len(colors)], edgecolor='black')
        axes[i].set_title(f'{column}', fontweight='bold')
        axes[i].set_xlabel('Value')
        axes[i].set_ylabel('Frequency')
        axes[i].grid(True, alpha=0.3)

        # Add mean line
        mean_val = new_df[column].mean()
        axes[i].axvline(mean_val, color='red', linestyle='--', alpha=0.8, label=f'Mean: {mean_val:.2f}')
        axes[i].legend()

plt.tight_layout()
plt.show()
print("Histograms created successfully!")
```

```python
# Task 2.5: Create pairplot visualization
print("Creating pairplot visualization...")

plt.figure(figsize=(12, 10))
pairplot = sns.pairplot(new_df, height=2.5, aspect=1.1,
                       plot_kws={'alpha': 0.6, 's': 20},
                       diag_kws={'alpha': 0.7})

pairplot.fig.suptitle('Pairplot of Au Nanoparticle Features', y=1.02, fontsize=16)

plt.show()
print("Pairplot completed successfully!")
```

```python
# Task 2.6: Custom PairGrid with specific plot types
print("Creating custom PairGrid visualization...")

# Create PairGrid
g = sns.PairGrid(new_df, height=3)

# Upper triangle: bivariate histograms
g.map_upper(sns.histplot, alpha=0.7)

# Diagonal: histograms with kernel density estimation (KDE)
g.map_diag(sns.histplot, kde=True, alpha=0.7)

# Lower triangle: bivariate kernel density estimation plots
g.map_lower(sns.kdeplot, fill=True, alpha=0.7)

# Add title
plt.suptitle('Custom PairGrid: Au Nanoparticle Dataset Analysis', y=1.02, fontsize=16, fontweight='bold')

plt.tight_layout()
plt.show()
print("Custom PairGrid completed successfully!")

# Additional analysis
print("\n" + "=" * 60)
print("ADDITIONAL DATASET INFORMATION:")
print("=" * 60)
print(f"Dataset shape: {new_df.shape}")
print(f"Total samples: {len(new_df)}")
print(f"Features analyzed: {len(columns_to_keep)}")
print(f"Missing values per column:")
for col in columns_to_keep:
    missing = new_df[col].isnull().sum()
    print(f"  {col}: {missing} ({missing/len(new_df)*100:.2f}%)")
```
